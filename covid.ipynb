```python
# Goal : Make a Ai Model to predict the spread of COVID-19

import pandas as pd
import numpy as np  
covid = pd.read_csv('covid_dataset.csv')
covid.head()
```

```python
# Understanding the dataset

## Display the Unique values in each column
distinct_values = {col: covid[col].unique() for col in covid.columns if col not in ['Age', 'Date_of_Death']}
distinct_values
```

```python
## Display the data types and non-null counts of each column
covid.info()
```

```python
## Displat the statistical summary of the dataset
covid.describe()
```

```python
# Data Preprocessing

## replace 97,98 and 99 with NaN
cols_to_clean = [
    'Intubated', 'Pneumonia', 'Pregnant', 'Diabetes', 'COPD', 
    'Asthma', 'Immunosuppressed', 'Hypertension', 'Other_Disease', 
    'Cardiovascular_Disease', 'Obesity', 'Chronic_Renal_Disease', 
    'Tobacco_Use', 'ICU_Admission'
]

for col in cols_to_clean:
    covid[col] = covid[col].replace({97: np.nan, 98: np.nan, 99: np.nan})
```

```python
## Drop the duplicated rows
covid.drop_duplicates(inplace = True)
```

```python
covid.info()
```

```python
cols = [
    'Intubated', 'Pneumonia', 'Age', 'Pregnant', 'Diabetes', 'COPD', 
    'Asthma', 'Immunosuppressed', 'Hypertension', 'Other_Disease', 
    'Cardiovascular_Disease', 'Obesity', 'Chronic_Renal_Disease', 
    'Tobacco_Use', 'ICU_Admission'
]
print(len(cols))
for col in cols:
    if covid[col].isnull().mean() < 0.2:  # if less than 20% missing
        covid[col].fillna(covid[col].mode()[0], inplace=True)
    else:
        print(f"Too many NaNs in {col}, consider dropping it or treating carefully.")
```
