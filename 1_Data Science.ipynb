```python
import pandas as pd
import numpy as np
import os

# Load the dataset
covid = pd.read_csv('./Datasets/covid_dataset.csv')

# Display the first few rows to understand the structure
covid.head()
```

```python
## Display the Unique values in each column
distinct_values = {col: covid[col].unique() for col in covid.columns if col not in ['Age', 'Date_of_Death']}
print(distinct_values)
```

```python
## Display the data types and non-null counts of each column
covid.info()
```

```python
# Display statistical summary of the dataset
covid.describe()
```

```python
# Replace specific values (97, 98, 99) with NaN for the specified columns
cols_to_clean = [
    'Intubated', 'Pneumonia', 'Pregnant', 'Diabetes', 'COPD', 
    'Asthma', 'Immunosuppressed', 'Hypertension', 'Other_Disease', 
    'Cardiovascular_Disease', 'Obesity', 'Chronic_Renal_Disease', 
    'Tobacco_Use', 'ICU_Admission'
]

for col in cols_to_clean:
    covid[col] = covid[col].replace({97: np.nan, 98: np.nan, 99: np.nan})
```

```python
# Drop duplicate rows if any
covid = covid.drop_duplicates()
```

```python
covid.info()
```

```python
# Check for any missing values in the dataset
print(covid.isnull().sum())
```

```python
cols = [
    'Intubated', 'Pneumonia', 'Age', 'Pregnant', 'Diabetes', 'COPD', 
    'Asthma', 'Immunosuppressed', 'Hypertension', 'Other_Disease', 
    'Cardiovascular_Disease', 'Obesity', 'Chronic_Renal_Disease', 
    'Tobacco_Use', 'ICU_Admission'
]
for col in cols:
    if covid[col].isnull().mean() < 0.2:  # if less than 20% missing
        covid[col].fillna(covid[col].mode()[0], inplace=True)
    else:
        # Remove rows where critical columns have missing values
        covid.drop(col, axis=1, inplace=True)

#Removed : [Intubated , Pregnant , ICU_Admission]

```

```python
# Convert 'Date_of_Death' column to datetime type
covid['Date_of_Death'] = pd.to_datetime(covid['Date_of_Death'], errors='coerce')
```

```python


# Create a new column 'is_dead' based on 'Date_of_Death'
covid['is_dead'] = covid['Date_of_Death'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
# Filter rows where 'Age' is positive, as negative or zero age doesn't make sense
covid = covid[covid['Age'] > 0]
```

```python
# Show the cleaned data
covid.head()
```

```python
# as in this dataset "1" means "yes" and "2" means "no", we will convert them to binary
binary_cols = [
     'Pneumonia', 'Diabetes', 'COPD', 
    'Asthma', 'Immunosuppressed', 'Hypertension', 'Other_Disease', 
    'Cardiovascular_Disease', 'Obesity', 'Chronic_Renal_Disease', 
    'Tobacco_Use','Patient_Care_Type' ,'Sex' , 'Treatment_Level'
]
for col in binary_cols:
    covid[col] = covid[col].replace({ 2: 0})  # Convert to bipolar values (1 and 0)

covid.head()
```

```python
# Display the Unique values in each column after cleaning
distinct_values = {col: covid[col].unique() for col in covid.columns if col not in ['Age', 'Date_of_Death']}
print(distinct_values)
```

```python
# grouping the data by COVID_Test_Result as 1-3 mean is_infected and 4+ mean is not infected
covid['is_infected'] = covid['COVID_Test_Result'].apply(lambda x: 1 if x <=3 else 0)
```

```python
# remove_irrelevant_cols for the model 
covid.drop(['COVID_Test_Result', 'Date_of_Death'], axis=1, inplace=True)
```

```python
# Normalize Age column to a range of 0-1
covid['Age'] = (covid['Age'] - covid['Age'].min()) / (covid['Age'].max() - covid['Age'].min())
```

```python
covid.describe()
```

```python
# export the cleaned dataset to a new CSV file for the ai model
# prevent Overwrite
file_path = './Datasets/cleaned_covid_dataset.csv'
base_name, extension = os.path.splitext(file_path)
counter = 1

while os.path.exists(file_path):
    file_path = f"{base_name}_{counter}{extension}"
    counter += 1

covid.to_csv(file_path, index=False)
print(f"Cleaned dataset saved at '{file_path}'")
```
